```python
labels = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio']
atributo = 6
```

```python
# Agrupamento dos dados em um só valor
## Média
media = []
media.append(dataset_jan.iloc[:,atributo].mean())
media.append(dataset_fev.iloc[:,atributo].mean())
media.append(dataset_mar.iloc[:,atributo].mean())
media.append(dataset_abr.iloc[:,atributo].mean())
media.append(dataset_mai.iloc[:,atributo].mean())
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
sns.set_palette("BuGn_r")

ax = sns.boxplot(data=[dataset_jan.iloc[:,atributo], dataset_fev.iloc[:,atributo],
                       dataset_mar.iloc[:,atributo], dataset_abr.iloc[:,atributo], dataset_mai.iloc[:,atributo]])
ax.plot(media, 'r', linewidth=2, label='Média')
ax.set_xticklabels(labels,
                    rotation=0, fontsize=10)
ax.set_title(dataset_jan.columns[atributo])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
sns.set_palette("BuGn_r")

ax = sns.violinplot(data=[dataset_jan.iloc[:,atributo], dataset_fev.iloc[:,atributo],
                       dataset_mar.iloc[:,atributo], dataset_abr.iloc[:,atributo], dataset_mai.iloc[:,atributo]])
ax.plot(media, 'r', linewidth=2, label='Média')
ax.set_xticklabels(labels,
                    rotation=0, fontsize=10)
ax.set_title(dataset_jan.columns[atributo])
```

```python
# Define valores de ferro (y) para cada mês
ferro=pd.concat([dataset_jan.iloc[:,21], dataset_fev.iloc[:,21],
                       dataset_mar.iloc[:,21], dataset_abr.iloc[:,21], dataset_mai.iloc[:,21]], axis=1)
```

```python
pair_jan = pd.concat([dataset_jan.iloc[:,atributo:atributo+1],ferro.iloc[:,0:1]], axis=1)
pair_fev = pd.concat([dataset_fev.iloc[:,atributo:atributo+1],ferro.iloc[:,1:2]], axis=1)
pair_mar = pd.concat([dataset_mar.iloc[:,atributo:atributo+1],ferro.iloc[:,2:3]], axis=1)
pair_abr = pd.concat([dataset_abr.iloc[:,atributo:atributo+1],ferro.iloc[:,3:4]], axis=1)
pair_mai = pd.concat([dataset_mai.iloc[:,atributo:atributo+1],ferro.iloc[:,4:]], axis=1)
pair_jan.columns=['Janeiro','Ferro']
pair_fev.columns=['Fevereiro','Ferro']
pair_mar.columns=['Março','Ferro']
pair_abr.columns=['Abril','Ferro']
pair_mai.columns=['Maio','Ferro']
```

```python
#fig, ax = plt.subplots(1, 1, figsize=(8, 5))
sns.set_palette("BuGn_r")

sns.pairplot(pair_jan, y_vars=pair_jan.columns[0], x_vars=pair_jan.columns[-1], kind='reg', height=3.5)
sns.pairplot(pair_fev, y_vars=pair_fev.columns[0], x_vars=pair_fev.columns[-1], kind='reg', height=3.5)
#ax.plot(media, 'r', linewidth=2, label='Média')
#ax.set_xticklabels(labels,
#                    rotation=0, fontsize=10)
#ax.set_title(dataset_jan.columns[atributo])
```
