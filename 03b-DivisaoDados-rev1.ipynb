```python
# Seleciona duas colunas, formando um agrupamento com os dados de um parâmetro
nivel_cx_alimentacao = cx_alimentacao.iloc[:,[0,1]]
densidade_cx_alimentacao = cx_alimentacao.iloc[:,[2,3]]
vazao_vol_cx_alimentacao = cx_alimentacao.iloc[:,[4,5]]
vazao_mas_cx_alimentacao = cx_alimentacao.iloc[:,[6,7]]
vazao_amina_cx_alimentacao = cx_alimentacao.iloc[:,[8,9]]
vazao_amido_cx_alimentacao = cx_alimentacao.iloc[:,[10,11]]

# Retira os dados nulos
selecao = nivel_cx_alimentacao.iloc[:,0] != 0
nivel_cx_alimentacao = nivel_cx_alimentacao[selecao]

selecao = densidade_cx_alimentacao.iloc[:,0] != 0
densidade_cx_alimentacao = densidade_cx_alimentacao[selecao]

selecao = vazao_vol_cx_alimentacao.iloc[:,0] != 0
vazao_vol_cx_alimentacao = vazao_vol_cx_alimentacao[selecao]

selecao = vazao_mas_cx_alimentacao.iloc[:,0] != 0
vazao_mas_cx_alimentacao = vazao_mas_cx_alimentacao[selecao]

selecao = vazao_amina_cx_alimentacao.iloc[:,0] != 0
vazao_amina_cx_alimentacao = vazao_amina_cx_alimentacao[selecao]

selecao = vazao_amido_cx_alimentacao.iloc[:,0] != 0
vazao_amido_cx_alimentacao = vazao_amido_cx_alimentacao[selecao]
```

```python
# Remove os dados que estão fora do intervalo
selecao = (nivel_cx_alimentacao.iloc[:,0] > base) & (nivel_cx_alimentacao.iloc[:,0] <= fim)
nivel_cx_alimentacao = nivel_cx_alimentacao[selecao]

selecao = (densidade_cx_alimentacao.iloc[:,0] > base) & (densidade_cx_alimentacao.iloc[:,0] <= fim)
densidade_cx_alimentacao = densidade_cx_alimentacao[selecao]

selecao = (vazao_vol_cx_alimentacao.iloc[:,0] > base) & (vazao_vol_cx_alimentacao.iloc[:,0] <= fim)
vazao_vol_cx_alimentacao = vazao_vol_cx_alimentacao[selecao]

selecao = (vazao_mas_cx_alimentacao.iloc[:,0] > base) & (vazao_mas_cx_alimentacao.iloc[:,0] <= fim)
vazao_mas_cx_alimentacao = vazao_mas_cx_alimentacao[selecao]

selecao = (vazao_amina_cx_alimentacao.iloc[:,0] > base) & (vazao_amina_cx_alimentacao.iloc[:,0] <= fim)
vazao_amina_cx_alimentacao = vazao_amina_cx_alimentacao[selecao]

selecao = (vazao_amido_cx_alimentacao.iloc[:,0] > base) & (vazao_amido_cx_alimentacao.iloc[:,0] <= fim)
vazao_amido_cx_alimentacao = vazao_amido_cx_alimentacao[selecao]
```

```python
# Zera o índice
nivel_cx_alimentacao.reset_index(drop=True, inplace=True)
densidade_cx_alimentacao.reset_index(drop=True, inplace=True)
vazao_vol_cx_alimentacao.reset_index(drop=True, inplace=True)
vazao_mas_cx_alimentacao.reset_index(drop=True, inplace=True)
vazao_amina_cx_alimentacao.reset_index(drop=True, inplace=True)
vazao_amido_cx_alimentacao.reset_index(drop=True, inplace=True)
```

```python
# Cria coluna para separação em grupos
nivel_cx_alimentacao['gr'] = -1
densidade_cx_alimentacao['gr'] = -1
vazao_vol_cx_alimentacao['gr'] = -1
vazao_mas_cx_alimentacao['gr'] = -1
vazao_amina_cx_alimentacao['gr'] = -1
vazao_amido_cx_alimentacao['gr'] = -1
```

```python
i = 0
base = bkp_base
for amostra in target.iloc[:,0]:
    base = amostra - np.timedelta64(2, 'h')
    
    # Critério de seleção dos dados com base na data e hora
    selecao = (nivel_cx_alimentacao.iloc[:,0] > base) & (nivel_cx_alimentacao.iloc[:,0] <= amostra)
    
    nivel_cx_alimentacao.loc[selecao, 'gr'] = i
    
    # Critério de seleção dos dados com base na data e hora
    selecao = (densidade_cx_alimentacao.iloc[:,0] > base) & (densidade_cx_alimentacao.iloc[:,0] <= amostra)
    
    densidade_cx_alimentacao.loc[selecao, 'gr'] = i
    
    # Critério de seleção dos dados com base na data e hora
    selecao = (vazao_vol_cx_alimentacao.iloc[:,0] > base) & (vazao_vol_cx_alimentacao.iloc[:,0] <= amostra)
    
    vazao_vol_cx_alimentacao.loc[selecao, 'gr'] = i
    
    # Critério de seleção dos dados com base na data e hora
    selecao = (vazao_mas_cx_alimentacao.iloc[:,0] > base) & (vazao_mas_cx_alimentacao.iloc[:,0] <= amostra)
    
    vazao_mas_cx_alimentacao.loc[selecao, 'gr'] = i
    
    # Critério de seleção dos dados com base na data e hora
    selecao = (vazao_amina_cx_alimentacao.iloc[:,0] > base) & (vazao_amina_cx_alimentacao.iloc[:,0] <= amostra)
    
    vazao_amina_cx_alimentacao.loc[selecao, 'gr'] = i
    
    # Critério de seleção dos dados com base na data e hora
    selecao = (vazao_amido_cx_alimentacao.iloc[:,0] > base) & (vazao_amido_cx_alimentacao.iloc[:,0] <= amostra)
    
    vazao_amido_cx_alimentacao.loc[selecao, 'gr'] = i
    
    if i % 5 == 0:
        print(i)
    i += 1
```
